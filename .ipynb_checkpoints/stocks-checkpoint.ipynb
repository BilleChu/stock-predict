```python
%matplotlib inline
import tensorflow as tf
import keras
import matplotlib.pylab as plt
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import TimeDistributed
from keras.layers import Dense
```

```python
import tushare as ts

df = ts.get_hist_data('600848')

ts.get_hist_data('600848',ktype='W') #获取周k线数据

ts.get_hist_data('600848',ktype='M') #获取月k线数据

ts.get_hist_data('600848',ktype='5') #获取5分钟k线数据

ts.get_hist_data('600848',ktype='15') #获取15分钟k线数据

ts.get_hist_data('600848',ktype='30') #获取30分钟k线数据

ts.get_hist_data('600848',ktype='60') #获取60分钟k线数据

# sh= ts.get_hist_data('sh')
# #ts.get_hist_data('sh')#获取上证指数k线数据，其它参数与个股一致，下同

# #ts.get_hist_data('sz')#获取深圳成指k线数据 ts.get_hist_data('hs300'）#获取沪深300指数k线数据

# #ts.get_hist_data('sz50')#获取上证50指数k线数据

# #ts.get_hist_data('zxb')#获取中小板指数k线数据

# #ts.get_hist_data('cyb')#获取创业板指数k线数
```

```python
df_1 = df.values[:, 1][np.newaxis]
sh_1 = df.values[:, 3][np.newaxis]
data = np.concatenate((df_1, sh_1), axis=0)
data = np.transpose(data)
print(data.shape)
```

```python
print('high price = ', df.values[:, 1][:10])
print('low price = ', df.values[:, 3][:10])
```

```python
data = np.flip(data, axis=0)
print(data[-1,:])
```

```python
validation = data[-10:, :]
train = data[:-10, :]
train_xx = np.ones([train.shape[0] - 10, 10, 2])
for i in range(train_xx.shape[0]):
    train_xx[i,:, :] = train[i:i+10, :]
print(train_xx[-1:, :, :])
label = data[10:-10, :]
label = label.reshape(label.shape[0], 1, 2)
print(label[-1:, :, :])
```

```python
BATCH_SIZE = 32
TIME_STEPS = 10
INPUT_SIZE = 2
CELL_SIZE = 15


model = Sequential()
model.add(LSTM(input_dim=2, output_dim=CELL_SIZE, return_sequences=True, stateful=False,))
OUTPUT_SIZE = 2
model.add(Dense(OUTPUT_SIZE, activation='linear'))
model.compile(optimizer='adam', loss='mse',)
```

```python
from keras.utils.visualize_util import plot
plot(model=model)
```

```python
history  = model.fit(train_xx, label, verbose=1, nb_epoch=300, batch_size=10)
```

```python
validation = data[-11:, :]
validation_xx = np.ones([validation.shape[0] - 10, 10, validation.shape[1]])
for i in range(validation_xx.shape[0]):
    validation_xx[i,:, :] = validation[i:i+10, :]
print(validation_xx.shape)
validation_xx = validation_xx.reshape((validation_xx.shape[0], validation_xx.shape[1], 2))
print(validation_xx.shape)
```

```python
pred = model.predict(train_xx)
print(pred.shape)
```

```python
y = pred[:, 0, :]
high_pred = y[:, 0]
low_pred = y[:, 1]
high_real = train[10:, 0]
low_real = train[10:, 1]
print(high_real.shape, high_pred.shape)
```

```python
plt.plot(high_real, label="high_real")
plt.plot(high_pred, label="high_pred")
plt.legend()
plt.show()
```

```python
plt.plot(low_real, label='low_real')
plt.plot(low_pred, label='low_pred')
plt.legend()
plt.show()
```
